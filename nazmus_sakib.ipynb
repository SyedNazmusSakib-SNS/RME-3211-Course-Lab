<div style="border-radius:4px; border-left:5px solid #DAA520; padding: 0px; background-color: #FFFAF0; font-size:100%; margin-bottom: 20px;">
  <h2 style="color: #8B4513; margin-top: 0;">1. Imports</h2>
  <p style="color: #000000;">In this section, we import all necessary libraries for our bias-variance tradeoff analysis.</p>
</div>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, r2_score
```

<a id="data_loading"></a>
<div style="border-radius:10px; border-left:5px solid #DAA520; padding: 15px; background-color: #FFFAF0; font-size:100%; margin-bottom: 20px;">
  <h2 style="color: #8B4513; margin-top: 0;">2. Data Loading</h2>
  <p style="color: #000000;">Here I load the dataset that will be used for our polynomial regression models.</p>
</div>

```python
df = pd.read_csv('boston_housing.csv')
```

```python
df.head()
```

```python
df.columns.to_list()
```

- Column names are too cryptic
- converting them into meaningful column names

```python
# Define a dictionary mapping old column names to new ones
new_column_names = {
    'id': 'Property_ID',
    'CRIM': 'Crime_Rate',
    'ZN': 'Residential_Zoning',
    'INDUS': 'Industrial_Pct',
    'CHAS': 'Charles_River_Access',
    'NOX': 'Nitrogen_Oxides',
    'RM': 'Avg_Rooms',
    'AGE': 'Pre_War_Built_Pct',
    'DIS': 'Employment_Center_Dist',
    'RAD': 'Highway_Access',
    'TAX': 'Property_TaxRate',
    'PTRATIO': 'Pupil_Teacher_Ratio',
    'B': 'Black_Residents_Pct',
    'LSTAT': 'Lower_Status_Pct',
    'class': 'Price'
}

df.rename(columns=new_column_names, inplace=True)

print(df.head())
```

<a id="train_test_split"></a>
<div style="border-radius:10px; border-left:5px solid #DAA520; padding: 15px; background-color: #FFFAF0; font-size:100%; margin-bottom: 20px;">
  <h2 style="color: #8B4513; margin-top: 0;">3. Train-Test Split</h2>
  <p style="color: #000000;">I split our dataset into training and testing sets to evaluate our models properly.</p>
</div>

```python
X = df.drop(['Price', 'Property_ID'], axis=1)
y = df['Price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

<a id="modeling"></a>
<div style="border-radius:10px; border-left:5px solid #DAA520; padding: 15px; background-color: #FFFAF0; font-size:100%; margin-bottom: 20px;">
  <h2 style="color: #8B4513; margin-top: 0;">4. Modeling</h2>
  <p style="color: #000000;">In this section, I created polynomial regression models of different degrees to observe the bias-variance tradeoff.</p>
</div>

<a id="poly_degree_1"></a>
<div style="border-radius:10px; border-left:5px solid #EEE8AA; padding: 15px; background-color: #FFFAF0; font-size:100%; margin-left: 30px; margin-bottom: 20px;">
  <h3 style="color: #8B4513; margin-top: 0;">4.1 Polynomial Degree 1</h3>
  <p style="color: #000000;">Linear regression model using polynomial features of degree 1.</p>
</div>

```python
# Degree 1 polynomial features (i.e., linear regression)
poly_1 = PolynomialFeatures(degree=1, include_bias=False)
X_train_poly_1 = poly_1.fit_transform(X_train)
X_test_poly_1 = poly_1.transform(X_test)
```

```python
# Fit linear regression
linreg_1 = LinearRegression()
linreg_1.fit(X_train_poly_1, y_train)
```

```python
# Intercept and coefficients
print("Intercept:", linreg_1.intercept_)
print("\nCoefficients:")
for name, coef in zip(X.columns, linreg_1.coef_):
    print(f"{name:25s}: {coef:.4f}")
```

```python
# Feature importance (absolute value of coefficients)
feature_importance = pd.Series(np.abs(linreg_1.coef_), index=X.columns)
top10 = feature_importance.sort_values(ascending=False).head(10)
print("\nTop 10 Feature Importances:")
print(top10)
```

```python
# Predictions
y_train_pred_1 = linreg_1.predict(X_train_poly_1)
y_test_pred_1 = linreg_1.predict(X_test_poly_1)

# Metrics
train_rmse_1 = root_mean_squared_error(y_train, y_train_pred_1)
test_rmse_1 = root_mean_squared_error(y_test, y_test_pred_1)
train_r2_1 = r2_score(y_train, y_train_pred_1)
test_r2_1 = r2_score(y_test, y_test_pred_1)

print(f"\nTrain RMSE for Degree 1: {train_rmse_1:.4f}")
print(f"Test RMSE for Degree 1:  {test_rmse_1:.4f}")
print(f"Train R2 for Degree 1:   {train_r2_1:.4f}")
print(f"Test R2 for Degree 1:    {test_r2_1:.4f}")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_test_pred_1, alpha=0.7, color='teal')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Price (Degree 1)')
plt.grid(True)
plt.show()
```
